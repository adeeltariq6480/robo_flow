# Temporary Reference Product Labeling
Return to the Label Tool, copy its Config URL, paste below, then use **Runtime → Run all**. No dataset or reference-product database records are created.

```
CONFIG_URL = input('Paste Config URL from Label Tool: ').strip()
assert CONFIG_URL.startswith('http'), 'Config URL is required'
```

```
import json, os, subprocess, sys, urllib.request
with urllib.request.urlopen(CONFIG_URL, timeout=60) as r: cfg=json.loads(r.read().decode())
for env,key in [('SUPABASE_URL','supabase_url'),('SUPABASE_SERVICE_ROLE_KEY','supabase_service_role_key'),('HF_TOKEN','hf_token'),('HF_DATASET_REPO','hf_dataset_repo'),('HF_MODEL_REPO','hf_model_repo')]: os.environ[env]=cfg.get(key,'')
os.environ['DEPLOY_TARGET']='colab'; os.environ['LOW_MEMORY_MODE']='false'; os.environ['HF_HUB_DISABLE_XET']='1'
repo=cfg['repo_url']
subprocess.run(['git','clone',repo,'robo_flow'], check=False)
os.chdir('/content/robo_flow/worker')
subprocess.run([sys.executable,'-m','pip','install','-q','-r','requirements.txt'], check=True)
subprocess.run([sys.executable,'scripts/colab_label_tool.py','--config-url',CONFIG_URL], check=True)
print('Complete — return to Label Tool to review and download.')
```
