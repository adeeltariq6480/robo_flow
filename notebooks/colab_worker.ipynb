# Robo Flow — Colab auto-label worker

Railway API creates `queued` jobs. This notebook **polls Supabase** and runs YOLO on Colab GPU.

1. Set secrets in the next cell (or Colab Secrets)
2. Runtime → Change runtime type → **T4 GPU**
3. Runtime → **Run all**

```python
import os

# Paste Railway / Supabase / HF values (or use Colab Secrets with same names)
os.environ["SUPABASE_URL"] = "https://YOUR_PROJECT.supabase.co"
os.environ["SUPABASE_SERVICE_ROLE_KEY"] = "YOUR_SERVICE_ROLE_KEY"
os.environ["HF_TOKEN"] = "hf_..."
os.environ["HF_DATASET_REPO"] = "Adeel6480/robo_flow"
os.environ["HF_MODEL_REPO"] = "Adeel6480/robo_flow"
os.environ["HF_DATASET_REPO_TYPE"] = "dataset"
os.environ["HF_MODEL_REPO_TYPE"] = "model"
os.environ["WORKER_LOOP"] = "true"
os.environ["POLL_SECONDS"] = "10"
os.environ["RUN_AUTO_LABEL_WORKER"] = "true"

try:
    from google.colab import userdata
    for key in (
        "SUPABASE_URL", "SUPABASE_SERVICE_ROLE_KEY", "HF_TOKEN",
        "HF_DATASET_REPO", "HF_MODEL_REPO",
    ):
        try:
            val = userdata.get(key)
            if val:
                os.environ[key] = val
        except Exception:
            pass
except ImportError:
    pass

print("Env ready. Dataset repo:", os.environ.get("HF_DATASET_REPO"))
```
